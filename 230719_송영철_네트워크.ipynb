```python
import numpy as np
import networkx as nx
import pandas as pd

year = 2000

file = "IPC 쌍별 빈도수_"+str(year)+"년.xlsx"

f = pd.read_excel(file)

a=[]
b=[]
fre=[]


for i in f.iloc[:,0]:
    a.append(i)
for i in f.iloc[:,1]:
    b.append(i)
for i in f.iloc[:,2]:
    fre.append(i)

G = nx.Graph()

for i in range(len(fre)):
    G.add_edge(a[i], b[i], weight=int(fre[i]))


pos = nx.spring_layout(G)

labels = nx.get_edge_attributes(G, 'weight')

nx.draw_networkx_edge_labels(G, pos, edge_labels = labels)
nx.draw(G, pos, with_labels=True)    
```
