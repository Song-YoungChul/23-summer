```python
import pandas as pd

f = pd.read_csv("./센서_클리닝.csv")
f
```

```python
#1) 원하는 열 하나만 뽑아내기 (ex, 명칭)

df = f[["명칭(원문)"]]
df
```

```python
#2) 조건에 맞는 열 여러개 뽑아내기 (ex, 출원이라는 단어가 포함된 열만)

df = f[[item for item in f.columns if "출원" in item]]
df
```

```python
#3) 원하는 행 하나만 뽑아내기 (ex, US9210192)

f[f["번호"] == "KR19940016922U"]
```

```python
#4-1*) 조건에 맞는 행 여러개 뽑아내기 (ex, 출원인 국가가 KR, 발명자가 현대, 출원일이 2021년도)


f.loc[(df['출원인국가'] == 'KR') & (df['출원인'] == '삼성전자주식회사') & (df['출원일'].str.contains('2018'))]
```

```python
#4-2*) 복합조건 출원인이 2020년도 이후이면서도, CPC 코드 중 H04L을 포함하는 특허만 뽑아내시오

f["출원년도"] = [int(str(i)[:4]) for i in f['출원일']]

df.loc[(f['출원년도'] >= 2020) & (f['국제특허분류'].str.contains('B08B'))]
```

```python
#5) 조건에 맞는 행과 열 복합 조건으로 뽑아내기

df.loc[(f['출원년도'] >= 2020) & (f['국제특허분류'].str.contains('B08B'))][["출원인","출원인주소"]]
```

```python
#6) 두개의 열 합쳐서 새로운 열 만들기 (출원인 김선혜 + from + 출원인 국가를 합치기 > '김선혜 from US')

f["새로운열"] = f["출원인"] + " from "+f["출원인국가"]
f
```

```python
#7) 하나의 열에서 원하는 부분 뽑아내서 새로운 열 만들기 (등록일에서 등록 년도만 뽑아내기)

f["출원년도"] = [int(str(i)[:4]) for i in f['출원일']]
f
```

```python
#8) 특허 국제 특허 분류 처리하기 (리스트)

f["국제특허분류(리스트)"] = [i.split(", ") for i in f["국제특허분류"]]
f
```

```python
#9-1) 인용 정보 네트워크로 만들기 (번호 from node - 자국피인용특허to node) - 자국 피인용 있는 애들로

from_lst = []
to_lst = []

for i, key in enumerate(f["번호"]):
    if pd.notna(f.loc[i, "자국피인용특허"]):
        a = f.loc[i, "자국피인용특허"].split(",")

        for j in a:
            from_lst.append(key)
            to_lst.append(j)
print(len(from_lst), len(to_lst))

ddf = pd.DataFrame(zip(from_lst,to_lst))
ddf
```

```python
#9-2) 9-1에서 뽑아낸 특허들의 제목만 뽑아오기 (from 특허번호, from 특허 제목, to 특허번호, to 특허 제목) 네개의 열을 가진 dataframe 만들기

from_title = []
to_title = []

for i in from_lst:
    for idx, j in enumerate(f["번호"]):
        if i == j:
            from_title.append(f.loc[idx, "명칭(원문)"])
            
for i in to_lst:
    c = 0
    for idx, j in enumerate(f["번호"]):
        if i == j:
            c=1
            to_title.append(f.iloc[idx, "명칭(원문)"])
        if c == 0:
            to_title.append("nan")

            
ddf = pd.DataFrame(zip(from_lst,from_title,to_lst,to_title))
ddf
```
