```python
import time
from bs4 import BeautifulSoup
from selenium.webdriver import Chrome
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

comments = []
dates = []
rates=[]

f = pd.read_excel("뽑아낼거.xlsx")
f = f[[0]]
f = f.iloc[:50,:]
f
```

```python
title = []
l = len(f.iloc[:50,0])

for k, n in enumerate(f.iloc[:50,0]):
    url = "http://lps3.www.wisdomain.com.sproxy.dongguk.edu/Search/Workboard/AdvancedHome"
    print(str(k+1)+" / "+str(l))
    driver = webdriver.Chrome(executable_path = 'C:/Chrome_driver/chromedriver.exe')
    driver.set_window_size(1920,1080)
   # driver.set_window_size(1,1)
    driver.get(url)
    
    time.sleep(1)
    
    #
    driver.find_element_by_xpath('//*[@id="mainSearchTabBox"]/li[5]/a').click()
    driver.find_element_by_xpath('//*[@id="ScriptSearch_Searchbox"]').send_keys("NUM="+n)
    driver.find_element_by_xpath('//*[@id="ScriptFormPartial"]/div[1]/ul/li/button').click()
    
    time.sleep(1)

    title_tags = driver.find_elements(By.CLASS_NAME, "custom-highlight-area")

    for idx, i in enumerate(title_tags):
        if idx==2:
            title.append(i.text)
            print(i.text)
    
    driver.close()

title
```

```python
f["제목"] = title
f.to_excel("크롤링결과_송영철.xlsx")
```
