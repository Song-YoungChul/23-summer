```python
import pandas as pd

f = pd.read_csv("./센서_클리닝.csv")
f
```

```python
df = f[["일련번호","번호","명칭(원문)","요약(원문)","출원인","출원인국가"]]
df
```

```python
import re
from konlpy.tag import Kkma
kkma = Kkma()
```

```python

lst = []

for i in df.iloc[:,3]:
    a = re.sub('[|0-9.,:(){}|]+', '', str(i))
    a = re.sub('[|  |]+', ' ', a)
    a = kkma.nouns(a)
    lst.append(a)
lst
```

```python
processed_data = []
for i in lst:
    txt=[]
    for j in i:
        txt.append(j)
    txt.append("\n")
    processed_data.append(txt)
processed_data[:100]
```

```python
from gensim.models.ldamodel import LdaModel
from gensim.models.callbacks import CoherenceMetric
from gensim import corpora
from gensim.models.callbacks import PerplexityMetric

import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

dictionary = corpora.Dictionary(processed_data)
dictionary.filter_extremes(no_below=2, no_above=0.5)
corpus = [dictionary.doc2bow(text) for text in processed_data]

num_topics = 50
chunksize = 2000
passes = 20
iterations = 400
eval_every = None

temp = dictionary[0]
id2word = dictionary.id2token

model = LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every
)

top_topics = model.top_topics(corpus) #, num_words=20)

# Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)

#from pprint import pprint
#pprint(top_topics)

import numpy as np
topics=[]
for i in top_topics:
    topic = []
#     print(i)
    for j in i:
        topic.append(j)
# #         j = j.astype("int32")
#         for k in range(len(j)):
#             print(j[k][1])
    topics.append(topic)
# topics

a = pd.DataFrame(topics)

topics = []
for i in a.iloc[:,0]:
    topic = []
    for j in i:
        topic.append(j[1])
    topics.append(topic)
    
topic = pd.DataFrame(topics)
topic
```

```python
stop_words = ["실시", "실시예", "내의"]
```

```python
lst = []
for i in topics:
    for j in i:
        if len(j) >1:
            if j not in lst:
                if j not in stop_words:
                    lst.append(j)
lst
```

```python
a = pd.DataFrame(index=range(len(processed_data)),columns=lst)

for i, k in enumerate(processed_data):
    for n, j in enumerate(lst):
        if j in k:
            a.iloc[i,n] = 1
        else:
            a.iloc[i,n] = 0
a
```

```python
company = "삼성전자주식회사"
cp_num = list(df[df["출원인"]==company].iloc[:,0].items())
cp_lst = []
for i in cp_num:
    cp_lst.append(i[1]-1)
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(a, b):
    return dot(a,b)/(norm(a)*norm(b))
```

```python
other_lst = []
for i in range(len(a.iloc[:,0])):
    if i not in cp_lst:
        other_lst.append(i)
other_lst
```

```python
asdf = []
for i in cp_lst:
    asdf.append(list(a.iloc[i,:]))
cp = pd.DataFrame(asdf,columns=lst)
cp_avr=pd.DataFrame(cp.mean())
cp_avr
```

```python
cos_lst = []
for i in other_lst:
    cos_lst.append(float(cos_sim(cp_avr.iloc[:,0], a.iloc[i,:])))
cos_lst
```

```python
sorted_lst = sorted(cos_lst, reverse=True)
sorted_lst = sorted_lst[:5]
sorted_lst
```

```python
for i, j in enumerate(cos_lst):
    if j in sorted_lst:
        print(df.iloc[other_lst[i],:])
        print(f'유사도 : {j}')
        print("-------------------------------------------------------------------------------------------------")
```
