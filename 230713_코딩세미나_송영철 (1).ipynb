```python
!pip install networkx
import networkx as nx
import pandas as pd

def make_df(dic):
    key = []
    value=[]
    
    for k, v in dic.items():
        key.append(k)
        value.append(v)
    return key, value

f = pd.read_csv("./1.csv")
f

```

```python
f["fre"] = 1
```

```python
a=[]
b=[]
fre=[]
de=[]
cl=[]
be=[]
ei=[]



for i in f.iloc[:,0]:
    a.append(i)
for i in f.iloc[:,1]:
    b.append(i)
for i in f.iloc[:,2]:
    fre.append(i)

G = nx.Graph()

for i in range(len(fre)):
    G.add_edge(a[i], b[i], weight=int(fre[i]))

de = nx.degree_centrality(G)
dek, dev = make_df(de)

cl = nx.closeness_centrality(G)
clk, clv = make_df(cl)

be = nx.betweenness_centrality(G)
bek, bev = make_df(be)

ei = nx.eigenvector_centrality(G)
eik, eiv = make_df(ei)

#print(nx.closeness_centrality(G))

df = pd.DataFrame(zip(dek, dev, clk, clv, bek, bev, eik, eiv), columns=["degree", "index","closeness", "index", "betweenness", "index", "eigenvector", "index"])
df
```
